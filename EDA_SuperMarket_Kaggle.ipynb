# Univariate Analysis

L'analyse univariée en EDA (Exploratory Data Analysis) consiste à examiner les caractéristiques d'une variable individuellement, sans considérer leur relation avec d'autres variables. L'objectif principal est de comprendre la distribution et les caractéristiques d'une seule variable à la fois.

Voici les grandes étapes pour réaliser une analyse univariée sur un dataset :

1. **Comprendre les types de variables** : Identifiez le type de données de chaque variable (numérique, catégorielle, temporelle, etc.). Cela déterminera les méthodes d'analyse appropriées.

2. **Statistiques descriptives** : Calculez des statistiques descriptives telles que la moyenne, la médiane, l'écart type, le minimum, le maximum, etc., pour les variables numériques. Pour les variables catégorielles, vous pouvez compter les occurrences de chaque catégorie.

3. **Visualisation des distributions** : Utilisez des histogrammes pour représenter graphiquement la distribution des variables numériques. Pour les variables catégorielles, vous pouvez utiliser des diagrammes à barres pour afficher la fréquence de chaque catégorie.

4. **Analyse des valeurs manquantes** : Identifiez les valeurs manquantes dans les données et évaluez leur impact potentiel sur l'analyse.

5. **Gestion des valeurs aberrantes** : Identifiez et évaluez les valeurs aberrantes (outliers) dans les variables numériques. Déterminez s'il est nécessaire de les traiter ou de les laisser en fonction du contexte de l'analyse.

6. **Analyse de la distribution** : Examinez la distribution des variables numériques pour déterminer si elles suivent une distribution normale ou une autre distribution spécifique. Vous pouvez utiliser des tests statistiques ou des visualisations comme les qq-plots pour cela.

7. **Analyse de la tendance centrale et de la dispersion** : Explorez la tendance centrale (moyenne, médiane, mode) et la dispersion (écart type, plage interquartile) des variables numériques pour comprendre leur comportement global.

8. **Analyse des valeurs catégorielles** : Pour les variables catégorielles, examinez la répartition des catégories et identifiez les catégories dominantes ouentes.

En résumé, l'analyse univariée permet d'obtenir un aperçu approfondi des caractéristiques individuelles du dataset, ce qui est essentiel pour comprendre la structure des données et orienter les analyses plus avancées dans le processus d'exploration des données.équentes.

layer, stack, fill, dodges-unes des autres options :

# Bivariate Analysis

Bivariate analysis is a statistical method that involves analyzing and understanding the relationship or association between two different variables. It explores how changes in one variable might be related to changes in another, examining their connection or correlation to draw insights from their interactions.

## Test de Chi-deux

Une fois que nous avons la valeur du chi-deux, nous la comparons à une valeur critique de la distribution du chi-deux, en tenant compte du degré de liberté. Si la valeur calculée du chi-deux est plus grande que la valeur critique, nous rejetons l'hypothèse nulle selon laquelle les variables sont indépendantes.

En d'autres termes, une valeur de chi-deux élevée indique que les variables ne sont probablement pas indépendantes et qu'il existe une relation significative entre elles.

```python
from scipy.stats import chi2_contingency
```

```python
# Calcul de la corrélation entre les variables continues
correlation = df['Variable1'].corr(df['Variable2'])

# Test du Chi-deux pour les variables catégorielles
chi2, p, dof, expected = chi2_contingency(pd.crosstab(df['Variable1'], df['Variable2']))
```

# Plot Using Aggregated Functions

```python
result1 = df.groupby('Category')['Postal Code'].agg(['mean', 'max', 'min']).reset_index()
result2 = df.groupby('Category')['Sales'].agg(['mean', 'max', 'min']).reset_index()

fig, axes = plt.subplots(2, 3, figsize=(40, 16))

sns.barplot(ax=axes[0, 0],data=result1, x='Category', y='min')
axes[0, 0].set_title('Min Postal Code of Different Category')
axes[0, 0].set_xlabel('Category')
axes[0, 0].set_ylabel('Min')


sns.barplot(ax=axes[0, 1],data=result1, x='Category', y='mean')
axes[0, 1].set_title('Mean Postal Code of Different Category')
axes[0, 1].set_xlabel('Category')
axes[0, 1].set_ylabel('Mean')

sns.barplot(ax=axes[0, 2],data=result1, x='Category', y='max')
axes[0, 2].set_title('Max Postal Code of Different Category')
axes[0, 2].set_xlabel('Category')
axes[0, 2].set_ylabel('Max')

sns.barplot(ax=axes[1, 0],data=result2, x='Category', y='min')
axes[1, 0].set_title('Min Sales of Different Category')
axes[1, 0].set_xlabel('Category')
axes[1, 0].set_ylabel('Min')


sns.barplot(ax=axes[0, 1],data=result2, x='Category', y='mean')
axes[1, 1].set_title('Mean Sales of Different Category')
axes[1, 1].set_xlabel('Category')
axes[1, 1].set_ylabel('Mean')

sns.barplot(ax=axes[1, 2],data=result2, x='Category', y='max')
axes[1, 2].set_title('Max Sales of Different Category')
axes[1, 2].set_xlabel('Category')
axes[1, 2].set_ylabel('Max')

plt.tight_layout()
plt.show()
```
